```python
import pandas as pd
```

```python
df = pd.read_excel("../datasets/Student_Mobility_2013-14.xlsx")
```

```python
sending = df.groupby(["SendingCountry", "LevelOfStudy"]).count().Action.unstack().reset_index(level=0)
receiving = df.groupby(["ReceivingCountry", "LevelOfStudy"]).count().Action.unstack().reset_index(level=0)
```

```python
sending = sending.rename(columns={"SendingCountry": "country", "First Cycle": "sending_first", 
                                  "Second Cycle": "sending_second", "Short Cycle": "sending_short", 
                                  "Third Cycle": "sending_third"})
```

```python
receiving = receiving.rename(columns={"ReceivingCountry": "country", "First Cycle": "receiving_first", 
                                  "Second Cycle": "receiving_second", "Short Cycle": "receiving_short", 
                                  "Third Cycle": "receiving_third"})
```

```python
sending.join(receiving)
```
